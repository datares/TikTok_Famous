### Data Collecting function 
This function takes a list of users as an argument and uses the api to add video data to a list. With conjunction of the `data_cleaner()` function, the `get_vids()` function outputs a cleaned verson of the list conatining video data.

```python
def get_vids(users, n_videos = 25):
    vids = []
    for username in users:
        try:
            vids = vids + api.byUsername(username, count=n_videos)
        except:
            pass
    return data_cleaner(vids)
```

```python
top_users_vids = get_vids(users['username'])
to_csv(top_users_vids, name = 'top_users+vids')
```
